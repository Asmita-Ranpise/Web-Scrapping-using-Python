```python
import requests
from bs4 import BeautifulSoup
```

```python
url="https://www.makaan.com/pune-residential-property/buy-property-in-pune-city"
req=requests.get(url)
soup=BeautifulSoup(req.text,'html')
```

```python
soup
```

* we can scrap data now

```python
area=soup.find('td',{'class':'size'})
area.text
```

```python
area1=soup.findAll('td',{'class':'lbl'})
area1
```

```python
Area=[]
sq1=soup.findAll('td',{'class':'size'})
for s in sq1:
  Area.append(s.text.strip('/'))
print(Area)
```

* **Price**

```python
price=soup.find('div',attrs={'data-type':'price-link'}).span.text
price
```

```python
Price=[]
price11=soup.find_all('div',attrs={'data-type':'price-link'})
for p in price11:
  Price.append(p.text.strip())
print(Price)
```

* Price per Square Foot i.e /**sqft**

```python
ppsq=soup.find('td',attrs={'class':'lbl rate'}).text
ppsq
```

```python
RPSQ=[]
p1=soup.find_all('td',attrs={'class':'lbl rate'})
for  p in p1:
  RPSQ.append(p.text.strip('/ sq ft'))
print(RPSQ)
```

* No. of **BHK**

```python
b=soup.find('a',attrs={'class':'typelink'})
b.span.text.strip()
```

```python
BHK=[]
b1=soup.find_all('a',attrs={'class':'typelink'})
for b in b1:
  BHK.append(b.span.text.strip())
print(BHK)
```

* Location

```python
loc=soup.find('a',attrs={'class':'loclink'})
loc.span.text
```

```python
Location=[]
loc1=soup.find_all('a',attrs={'class':'loclink'})
for l in loc1:
  Location.append(l.span.text)
print(Location)
```

***Bath Room***

```python
Bath = []
bathroom = soup.find_all('li',attrs={'class':'keypoint','title':'Bathrooms'})
for b in bathroom:
  Bath.append(b.span.text.strip())
print(Bath)
```

 **Status**

```python
Status=[]
st=soup.find_all('td',attrs={'class':'val'})
for i in st:
  Status.append(i.text)
print(Status)
```

**Builder**

```python
Builder=[]
bt=soup.find_all('a',attrs={'class':'seller-name',})
for i in bt:
  Builder.append(i.span.text.strip())
print(Builder)
```

**All variables are taken in dictionary**

```python
prop_pune={'Price':Price,
           'Area in sqft':Area,
           'Rate per sqft':RPSQ,
           'No. of BHK':BHK,
           'Location':Location,
           'Status':Status,
           'Builder':Builder}
print(prop_pune)
```

* **To insert all data of all pages import time module and give url at last ?page=' as base_url**

importing time

```python
import time
from IPython.display import clear_output


base_url="https://www.makaan.com/pune-residential-property/buy-property-in-pune-city?page="

Price1=[]
Area1=[]
RPSQ1=[]
BHK1=[]
Location1=[]
Status1=[]
Builder1=[]


for i in range(1,26):
  print(f'Going to Scrap Data from the {i} page')
  clear_output(wait=True)
  time.sleep(2)
  url=base_url + str(i)
  #htps://wwtw.makaan.com/hyderabad-residential-property/buy-property-in-pune-city?page=
  #request the url and convert into soup object
  req=requests.get(url)
  soup=BeautifulSoup(req.text,'html')


  price11=soup.find_all('div',attrs={'data-type':'price-link'})
  for p in price11:
    Price1.append(p.text.strip())
  print(Price1)

  
  sq1=soup.findAll('td',{'class':'size'})
  for s in sq1:
    Area1.append(s.text.strip('/'))
  print(Area1)


  
  p1=soup.find_all('td',attrs={'class':'lbl rate'})
  for  p in p1:
    RPSQ1.append(p.text.strip('/ sq ft'))
  print(RPSQ1)


  
  loc1=soup.find_all('a',attrs={'class':'loclink'})
  for l in loc1:
    Location1.append(l.span.text)
  print(Location1)


  
  st=soup.find_all('td',attrs={'class':'val'})
  for i in st:
    Status1.append(i.text)
  print(Status1)


  
  bt=soup.find_all('a',attrs={'class':'seller-name',})
  for i in bt:
    Builder1.append(i.span.text.strip())
  print(Builder1)



  
  b1=soup.find_all('a',attrs={'class':'typelink'})
  for b in b1:
    BHK1.append(b.span.text.strip())
  print(BHK1)
    
```

```python
prop_info = {'Price' : Price1,
            'Rate_per_sqft' : RPSQ1,
            'Area' : Area1,
            'Num_Bhk' : BHK1,
            'Locality' : Location1,
            'Status' : Status1,
             'Builder' : Builder1
             }
print(prop_info)
```

```python
import pandas as pd
data = pd.DataFrame(prop_info)
data
```
