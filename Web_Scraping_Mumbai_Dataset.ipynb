* Firstly import requests
* Secondly import BeautifulSoup from bs4

```python
import requests
from bs4 import BeautifulSoup
```

* copy paste url in variable **url**
* get the url in req variable
* In soup variable type BeautifulSoup(req.text,'html')

```python
url = 'https://www.makaan.com/mumbai-residential-property/buy-property-in-mumbai-city'
req = requests.get(url)
soup = BeautifulSoup(req.text,'html')
```

```python
soup
```

* You can scrape data by **soup.find or soup.find_all** tag

```python
sq = soup.find('td',attrs={"class":"size"})
sq.text
```

* Area_sqft

```python
sqft1 = soup.find_all('td',attrs={'class':'size'})
sqft1
```

```python
area_sqft=[]
area_sqft1 = soup.find_all('td',attrs={'class':'size'})
for s in sqft1:
  area_sqft.append(s.text.strip())
print(area_sqft)
```

* price

```python
pr = soup.find('div',attrs= {'data-type': 'price-link'}).span.text
pr
```

```python
price=[]
pr1 = soup.find_all('div',attrs={'data-type':'price-link'})
for p in pr1:
  price.append(p.text.strip())
print(price)
```

* Sqft

```python
sq = soup.find('td',attrs={'class':'lbl rate'})
sq.text 
```

```python
sqft = []
sq1 = soup.find_all('td',attrs={'class':'lbl rate'})
for s1 in sq1:
  sqft.append(s1.text.strip('/sq ft'))
print(sqft)
```

```python
b = soup.find('a',attrs={'class':'typelink'})
b.span.text.strip()
```

* Bhk

```python
bhk = []
bh = soup.find_all('a',attrs={'class':'typelink'})
for b1 in bh:
  bhk.append(b1.span.text.strip())
print(bhk)
```

* Location

```python
location = []
loc = soup.find_all('a',attrs={'class':'loclink'})
for l in loc:
  location.append(l.span.text)
print(location) 
```

* Bathrooms

```python
bath = []
bathroom = soup.find_all('li',attrs={'class':'keypoint','title':'Bathrooms'})
for b in bathroom:
  bath.append(b.span.text.strip())
print(bath)
```

* Status

```python
status = []
st = soup.find_all('td',attrs={'class':'val'})
for s2 in st:
  status.append(s2.text)
print(status)
```

* Builder

```python
builder = []
build = soup.find_all('a',attrs={'class':'seller-name'})
for b3 in build:
  builder.append(b3.span.text.strip())
print(builder)
```

* All variables are taken it in a dictionary

```python
prop_mum = {'Price' : price,
            'Rate_per_sqft' : sqft,
            'Area' : area_sqft,
            'Num_Bhk' : bhk,
            'Locality' : location,
            'Status' : status,
            'Builder' : builder}

print(prop_mum)
```

* **To insert all data of all pages import time module and give url at last ?page=' as base_url**

```python
import time
from IPython.display import clear_output
```

```python
base_url = 'https://www.makaan.com/hyderabad-residential-property/buy-property-in-mumbai-city?page='

price1 = []
n_bhk = []
area = []
loc = []
rate_info = []
status1 = []

for i in range(1,26):

  print(f'Going to scrape the data from the {i} page')
  clear_output(wait=True)
  time.sleep(2)

  url = base_url+str(i)
  # 'https://www.makaan.com/hyderabad-residential-property/buy-property-in-hyderabad-city?page=' 

  #request the url and convert into soup object
  req = requests.get(url)
  soup = BeautifulSoup(req.text,'html')

  # price

  p_tag = soup.find_all('div',attrs={'data-type' : 'price-link'})
  for p in p_tag:
    price1.append(p.text.strip())


  # number of bhk
  prop=soup.find_all('a',attrs={'class':'typelink','target':'_blank'})
  for b in prop:
    n_bhk.append(b.span.text.strip())
  
  # area in sqft
  a_tag = soup.find_all('td', attrs ={'class':'size'})
  for a in a_tag:
    area.append(a.text.strip())
 
  # locality
  l_tag=soup.find_all('a',attrs={'class':'loclink'})
  for l in l_tag:
    loc.append(l.text.strip())

  # price_per_sqft
  rate_tag=soup.find_all('td',attrs={'class':'lbl rate'})
  rate=[r.text.strip() for r in rate_tag]
  for rt in rate:
    rate_info.append(rt)

  status_tag=soup.find_all('td',attrs={'class':'val'})
  for s in status_tag:
    status1.append(s.text.strip())
```

```python
prop_info = {'Price' : price1,
            'Rate_per_sqft' : rate_info,
            'Area' : area,
            'Num_Bhk' : n_bhk,
            'Locality' : loc,
            'Status' : status1}
print(prop_info)
```

```python
import pandas as pd
data = pd.DataFrame(prop_info)
data
```
